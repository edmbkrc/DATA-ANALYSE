```python
## CONDITIONAL STATEMENTS
```

```python
arr = np.random.randint(0,30,(10,3))
df = pd.DataFrame(arr)
df.columns = ["var1","var2","var3"]
df
```

```python
df[df["var1"] > 15]
```

```python
df[df["var1"] > 15]["var2"]
```

```python
df[df["var3"] < 10]["var2"]
```

```python
df
```

```python
df[(df["var2"] < 15) & (df["var3"] > 5) & (df["var1"] != 3)]
```

```python
df[(df["var2"] < 15) | (df["var3"] > 5) | (df["var1"] == 3)]
```

```python
df[(df["var2"] < 15) | (df["var3"] > 5) | (df["var1"] == 3)].iloc[2:5]
```

```python
df[(df["var2"] < 15) | (df["var3"] > 5) | (df["var1"] == 3)].iloc[2:5,[0,2]]
```
