```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Iris veri setini DataFrame'e oku
df = pd.read_csv('iris.csv')

# Grafik çizimi
plt.figure(figsize=(12, 6))

# Species ve Petal Width
plt.subplot(1, 2, 1)
sns.stripplot(x='species', y='petal_width', data=df)
plt.title('Species ve Petal Width İlişkisi')
plt.xlabel('Species')
plt.ylabel('Petal Width')

# Species ve Sepal Width
plt.subplot(1, 2, 2)
sns.stripplot(x='species', y='sepal_width', data=df)
plt.title('Species ve Sepal Width İlişkisi')
plt.xlabel('Species')
plt.ylabel('Sepal Width')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Iris veri setini DataFrame'e oku
df = pd.read_csv('iris.csv')

# Grafik çizimi
plt.figure(figsize=(10, 6))

# Species ve Petal Width
sns.scatterplot(x='petal_width', y='sepal_width', hue='species', data=df, palette='Set1')

plt.title('Species, Petal Width ve Sepal Width İlişkisi')
plt.xlabel('Petal Width')
plt.ylabel('Sepal Width')

plt.show()
```
